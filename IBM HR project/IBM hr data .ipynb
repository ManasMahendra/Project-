```python
# Task 6 - What is the highest percentage of Marital Status ?
# There are 45.78 % of employees are married .
```

```python
fig, ax = plt.subplots(figsize= (6,4))
ax = sb.countplot(x=data['MaritalStatus'])
for bar in ax.patches:                                             
    percentage =f"{round(bar.get_height() /data.shape[0]*100, 2)}%"         
                                                                    
    x = bar.get_x() + bar.get_width() /2                           
    y = bar.get_height()                                           
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" )
```

```python
# Task 7 - Which Job Role have the highest percentage of employees .
# There are 22.18 % of employees are in  Sales Executive job role .
```

```python
ax = sb.countplot(x=data['JobRole'])
for bar in ax.patches:                                             
    percentage =f"{round(bar.get_height() /data.shape[0]*100, 2)}%"         
                                                                    
    x = bar.get_x() + bar.get_width() /2                           
    y = bar.get_height()                                           
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" )
plt.xticks(rotation=90)
plt.show()
```
